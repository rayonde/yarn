```python
import numpy as np
from qutip import *
from yarn.PhilReinholdPygrape import *
from yarn.qutipHelpers import (
    plotWignersIntermediateStates, 
    plotOccupationsIntermediateStates,
    plotExpectation, cat
)
```

## parameters

```python
NQ = 2
NC = 20
drive = 1e-3 # GHz

# first order term:
chi = 3e-3 # GHz

# second order terms:
anharmonicityQ = .4 # GHz
kerrC = 1e-5 # GHz
```

## operators

```python
aC = tensor(qeye(NQ), destroy(NC))
aCd = aC.dag()

aQ = tensor(destroy(NQ), qeye(NC))
aQd = aQ.dag()
numQ = aQd * aQ
```

## drift Hamiltonian

```python
Hdrift = chi * aCd*aC * aQd*aQ
Hdrift+= (anharmonicityQ/2) * aQd * aQd * aQ * aQ
Hdrift+= (kerrC/2) * aCd * aCd * aC * aC
Hdrift*= 2*np.pi
```

## control Hamiltonian

```python
drive*= 2*np.pi
XQ =      drive * (aQ + aQd)
YQ = 1j * drive * (aQ - aQd)
XC =      drive * (aC + aCd)
YC = 1j * drive * (aC - aCd)
HcontrolList = [XQ, YQ, ]
```

## initial and target states

this is to measure the parity of photon number of cavity by obeserving the qubit


```python
inits, finals = [], []
for i in range(8):
    inits.append(  ket([0  , i], [NQ, NC]) )
    finals.append( ket([i%2, i], [NQ, NC]) )
```

## define the `loss_vec`

**TODO**: no explanation is found for this

```python
loss_vec = 1 - np.eye(NC)[-1]
loss_vec = np.kron(np.ones(2), loss_vec)
```

## define the penalties

```python
penalties = [
    make_lin_deriv_cost(5e-5),
    make_l1_penalty(1e-4, 1e2),
]
```

## define the grape setup

```python
setup = StateTransferSetup(
    Hdrift, HcontrolList, 
    inits, finals, 
    loss_vec=loss_vec, coherent=False, 
)
```

## define the initial control sequence

```python
initControls = 10 * random_waves(
    len(HcontrolList), 200, npoints=10
)
```

## run grape

```python
import warnings; warnings.filterwarnings('ignore')

result = run_grape(
    initControls, setup, 
    penalty_fns=penalties, term_fid=0.99 
)
```

## plot

```python
plot(result, ctrlNames=['XQ', 'YQ', ])
```

## evaluate the resulting pulse using `qutip`

```python
Ht = [list(a) for a in zip(HcontrolList, result.controls)]
Ht.append(Hdrift)
psi0 = ket([0, 1], [NQ, NC])
time = result.ts * 1.0

mesolveResult = mesolve(Ht,psi0,time,[],[])
```

## plot `qutip.mesolve` results  

```python
observables = [{'Hdrift':Hdrift, 'numQ':numQ}]
plotExpectation(time,observables,mesolveResult)
ptraceIndex = 0 # qubit:0, cavity:1
plotWignersIntermediateStates(
    time,mesolveResult,number=4,radius=4,
    ptraceIndex = ptraceIndex, 
)
plotOccupationsIntermediateStates(
    time,mesolveResult,number=4,
    ptraceIndex = ptraceIndex, 
)
```
